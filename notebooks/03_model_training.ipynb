# Base Model Training (Logistic Regression)

```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
                            confusion_matrix, 
                            f1_score,
                            accuracy_score,
                            precision_score,
                            recall_score
                            )
warnings.filterwarnings('ignore')
```

```python
seed = 42
np.random.seed(seed)
```

### Load The Data

```python
X_train = np.load('../artifacts/imbalance/X_train.npz')['arr_0']
Y_train = np.load('../artifacts/imbalance/Y_train.npz')['arr_0']
X_test = np.load('../artifacts/imbalance/X_test.npz')['arr_0']
Y_test = np.load('../artifacts/imbalance/Y_test.npz')['arr_0']


X_train_balance = np.load('../artifacts/balance/X_train.npz')['arr_0']
Y_train_balance = np.load('../artifacts/balance/Y_train.npz')['arr_0']
X_test_balance = np.load('../artifacts/balance/X_test.npz')['arr_0']
Y_test_balance = np.load('../artifacts/balance/Y_test.npz')['arr_0']
```

### Model Training 

```python
model_lr_imbalanced = LogisticRegression(
                            random_state = 42,
                            max_iter = 1000      
                            )
model_lr_imbalanced.fit(X_train, Y_train)
```

```python
model_lr_balanced = LogisticRegression(
                            random_state = 42,
                            max_iter = 1000      
                            )
model_lr_balanced.fit(X_train_balance, Y_train_balance)
```

### Inference

```python
Y_hat_train = model_lr_imbalanced.predict(X_train)
Y_hat_test = model_lr_imbalanced.predict(X_test)

Y_proba_train = model_lr_imbalanced.predict_proba(X_train)
Y_proba_test = model_lr_imbalanced.predict_proba(X_test)
```

```python
Y_hat_train_balanced = model_lr_balanced.predict(X_train_balance)
Y_hat_test_balanced = model_lr_balanced.predict(X_test_balance)

Y_proba_train_balanced = model_lr_balanced.predict_proba(X_train_balance)
Y_proba_test_balanced = model_lr_balanced.predict_proba(X_test_balance)
```

### Evaluation

```python
accuracy_imbalanced = accuracy_score(Y_test, Y_hat_test)
precision_imbalanced = precision_score(Y_test, Y_hat_test)
recall_imbalanced = recall_score(Y_test, Y_hat_test)
f1_score_imbalanced = f1_score(Y_test, Y_hat_test)
```

```python
accuracy_balanced = accuracy_score(Y_test_balance, Y_hat_test_balanced)
precision_balanced = precision_score(Y_test_balance, Y_hat_test_balanced)
recall_balanced = recall_score(Y_test_balance, Y_hat_test_balanced)
f1_score_balanced = f1_score(Y_test_balance, Y_hat_test_balanced)
```

```python
print(f"accuracy imbalanced : {accuracy_imbalanced}")
print(f"precision imbalanced : {precision_imbalanced}")
print(f"recall imbalanced : {recall_imbalanced}")
print(f"f1_score imbalanced : {f1_score_imbalanced}")
```

```python
print(f"accuracy balanced : {accuracy_balanced}")
print(f"precision balanced : {precision_balanced}")
print(f"recall balanced : {recall_balanced}")
print(f"f1_score balanced : {f1_score_balanced}")
```

```python
cm = confusion_matrix(Y_test, Y_hat_test)
sns.heatmap(cm, annot=True, fmt='d')
print("Confusion Matrix for Imbalanced Data")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

cm = confusion_matrix(Y_test_balance, Y_hat_test_balanced)
sns.heatmap(cm, annot=True, fmt='d')
print("Confusion Matrix for Balanced Data")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Save Logistic Regression Model

```python
import joblib

# Save the model to a file
joblib_file = "../models/logistic_regression_model_for_imbalanced.pkl"
joblib.dump(model_lr_imbalanced, joblib_file)

joblib_file = "../models/logistic_regression_model_for_balanced.pkl"
joblib.dump(model_lr_balanced, joblib_file)
```
